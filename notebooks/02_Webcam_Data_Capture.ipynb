```python
# this to test the camera and capture the video frame every 3 seconds

from anyio import current_time
import cv2
import time 
import os 

cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Cannot open camera")
    exit()

print("Camera opened successfully")

last_capture_time = time.time()
capture_interval = 3 # seconds
current_time = time.time()
caption_until = current_time
output_dir = os.path.join("..", "data", "raw", "20260427_dataset")

while True: 
    ret, frame = cap.read() # ret is a boolean indicate if the frame is read correctly
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # save the frame every 3 seconds
    current_time = time.time()
    if current_time - last_capture_time >= capture_interval:
        timestamp = time.strftime("%Y%m%d-%H%M%S")
        filename = f"frame_{timestamp}.jpg"
        output_path = os.path.join(output_dir, filename)
        cv2.imwrite(output_path, frame)
        print(f"Captured {output_path}")
        last_capture_time = current_time
        caption_text = f"Captured at {timestamp}"
        caption_until = current_time + 2 # display caption for 2 seconds
    
    # draw caption on frame if still within display window
    if time.time() < caption_until:
        cv2.putText(frame, caption_text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) == ord('q'):
        break   

cap.release()
cv2.destroyAllWindows()
```
